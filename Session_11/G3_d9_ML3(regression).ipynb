```python
#w0 - intercept
#w1 - slope
w0 = 1
w1 = 1.5

yh = w0 + w1*x
print(yh)
```

```python
error = ((y - yh)**2).mean()
print(error)
```

```python
dew0 = -2*(y-yh).mean()
print(dew0)
dew1 = -2*((y-yh)*x).mean()
print(dew1)
```

```python
lr = 0.01
w0 = w0 - lr * dew0
w1 = w1 - lr * dew1

print(w0 , w1)
```

```python
yh = w0 + w1*x
error = ((y - yh)**2).mean()
print(error)
```

```python
w0 = 1
w1 = 1.5

for i in range(500):
  yh = w0 + w1*x

  error = ((y - yh)**2).mean()

  dew0 = -2*(y-yh).mean()
  dew1 = -2*((y-yh)*x).mean()

  lr = 0.01
  w0 = w0 - lr * dew0
  w1 = w1 - lr * dew1

  print(error)
```

**Scikit-learn to implement linear regression**

```python
x = np.array([3,6,4,9,2]).reshape(5,1)   #Feature
y = np.array([7,8,6,10,5]).reshape(5,1)       #Target
```

```python
from sklearn.linear_model import LinearRegression      #Import algorithm
modelA = LinearRegression()                               #Define the model

modelA.fit(x,y)                                        #fit the data for training
```

```python
#Trained parameters
print("w1",modelA.coef_)
print("w0",modelA.intercept_)
```

```python
#Predictions made on x
yp = modelA.predict(x)
print(yp)
```

```python
4.05194805 + 0.65584416 * x
```

```python
error = abs(y - yp).mean()        #difference between real labels and predicted label => error
print(error)
```

```python
x_new = 5
y_new = modelA.predict([[x_new]])
print(y_new)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Build a machine learning model using following data to predict the sales value**

```python
df = pd.read_csv('https://github.com/bipulshahi/Dataset/raw/refs/heads/main/Advertising.csv',
                 index_col=0)
df.head()
```

**Simple linear regression**

```python
X = df[['TV']]          #TV advertisement cost
y = df['sales']          #Sales
```

```python
#build and evaluate sales prediction model using simple linear regression
```

```python
#Split the data into training and test sets - Evaluate model performance on data used for training and not used for training
from sklearn.model_selection import train_test_split
xtrain, xtest , ytrain, ytest = train_test_split(X,y,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
modelB = LinearRegression()

modelB.fit(xtrain,ytrain)
```

```python
#Average error while predicting for training data
ytrainP = modelB.predict(xtrain)
```

```python
#Average error while predictiong for test data
ytestP = modelB.predict(xtest)
```

```python
#Comapre predicted outcome (predicted sales) on train data with ytrain (real sales)
mae_train = abs(ytrain - ytrainP).mean()
print(mae_train)
```

```python
#Evaluste model performance on test data (data not used for training)
mae_test = abs(ytest - ytestP).mean()
print(mae_test)
```

```python
tv_expense = [[34.5]]
predicted_sales = modelB.predict(tv_expense)
print(predicted_sales)
```
