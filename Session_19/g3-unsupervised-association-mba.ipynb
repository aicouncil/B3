```python
df1 = df[df['Country'] == 'Germany']
df1.head()
```

```python
df1.dtypes
```

```python
df1.isna().sum()
```

```python
df1[df1['InvoiceNo'].astype(str).str.contains('C')]
```

```python
df2 = df1[~df1['InvoiceNo'].astype(str).str.contains('C')]
df2.head()
```

```python
df2['InvoiceNo'] = df2['InvoiceNo'].astype(float)
```

```python
df2.isna().sum()
```

```python
df3 = df2.drop(columns = 'lower')
df3.head()
```

```python
df3['Description'] = df3['Description'].str.lower().str.strip()
```

```python
df3.head()
```

```python
df4 = df3[df3['Description'] != 'postage']
df4.head()
```

```python
df_basket = pd.crosstab(df4['CustomerID'] , df4['Description'])
df_basket.head()
```

```python
df_basket = df_basket.apply(lambda x : x > 0)
df_basket.head()
```

```python
#Identify the most frequent itemsets

from mlxtend.frequent_patterns import apriori
frequent_itemsets = apriori(df_basket, use_colnames=True,min_support=0.08)
frequent_itemsets
```

```python
#Calculate confidence & Lift values for frequent itemsets

from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets , metric='lift' , min_threshold=1)

rules
```

```python
r1 = rules[['antecedents','consequents','support', 'confidence', 'lift']]
r1['antecedents'] = r1['antecedents'].apply(lambda x : list(x))
r1['consequents'] = r1['consequents'].apply(lambda x : list(x))
r1.to_csv('recommendar.csv')
```

```python
df_recommendor = pd.read_csv('/kaggle/working/recommendar.csv', index_col=0)
df_recommendor.head()
```

```python
df_recommendor = df_recommendor.sort_values('lift' , ascending = False)
```
