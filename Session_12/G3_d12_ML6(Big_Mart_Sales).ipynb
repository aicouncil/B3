```python
X = df2.drop(columns = ['Item_Outlet_Sales'])
y = df2['Item_Outlet_Sales']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
Xt = scaler.fit_transform(X)
```

```python
yt = np.log(df['Item_Outlet_Sales'])
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(Xt,yt)
```

```python
from sklearn.linear_model import LinearRegression
modelA = LinearRegression()
modelA.fit(xtrain,ytrain)
```

```python
ytrainP = modelA.predict(xtrain)
ytestP = modelA.predict(xtest)
```

```python
maeTrain = abs(ytrain - ytrainP).mean()
maeTest = abs(ytest - ytestP).mean()

print(f"Mean absolute error (Train) - {maeTrain}")
print(f"Mean absolute error (Test) - {maeTest}")
```

```python
n1 = ((ytrain - ytrainP)**2).sum()                 #Sum of square of difference between real sales and predicted sales
```

```python
d1 = ((ytrain - ytrain.mean())**2).sum()           #Sum of square of difference between given labels and average of given labels
```

```python
r2_score = (1- (n1/d1))
print(r2_score)
```

```python
from sklearn.metrics import r2_score

print("R2_Score_train" , r2_score(ytrain , ytrainP))
print("R2_Score_test" , r2_score(ytest , ytestP))
```
