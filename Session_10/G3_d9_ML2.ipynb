```python
import pandas as pd
import numpy as np
df = pd.read_csv('https://github.com/bipulshahi/Dataset/raw/refs/heads/main/wine.csv')
df.head()
```

```python
df['Target'].unique()
```

```python
X = df.drop(columns = ['Target'])
y = df['Target']
```

```python
X.head()
```

```python
from sklearn.preprocessing import MinMaxScaler      #Library to perform data scaling
scaler = MinMaxScaler()                            #Object to call scaling methods
scaler.fit(X)                                         #Fit the data to calculate metrics for scaling - Minimum and Maximum of X
Xscaled = scaler.transform(X)                                   #To perform final scaling
```

```python
Xscaled                             #All data columns have magnitude 0 to 1
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(Xscaled,y,random_state=2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
modelA = KNeighborsClassifier(n_neighbors=7)
modelA.fit(xtrain,ytrain)
```

```python
#Predicted output on train data
ytrainP = modelA.predict(xtrain)

#Predicted output on test data
ytestP = modelA.predict(xtest)
```

```python
print((ytrain == ytrainP).sum()/len(xtrain))
```

```python
print((ytest == ytestP).sum()/len(xtest))
```

```python
wine_feature = [13.24, 2.59, 2.87, 21.0, 118, 2.80, 2.69, 0.39, 1.82, 4.32, 1.04, 2.93, 735]
print(modelA.predict([wine_feature]))
```

**Cross-Validation**
  * K-fold cross validation - Change the data in train & test for k times and evaluate the performance each time

```python
train_accuracy = []
test_accuracy = []

for k in range(0,10):
  xtrain,xtest,ytrain,ytest = train_test_split(Xscaled,y,random_state=k)
  modelB = KNeighborsClassifier()
  modelB.fit(xtrain,ytrain)
  #print(modelB.score(xtrain,ytrain))
  #print(modelB.score(xtest,ytest))
  train_accuracy.append(modelB.score(xtrain,ytrain))
  test_accuracy.append(modelB.score(xtest,ytest))
```

```python
print("Average train accuracy (10-fold)-" , np.mean(train_accuracy))
print("Average test accuracy (10-fold)-" , np.mean(test_accuracy))
```
