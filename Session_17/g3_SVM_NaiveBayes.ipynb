```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)
df2 = imputer.fit_transform(df1)
```

```python
df3 = pd.DataFrame(df2 , columns = df1.columns)
df3.head()
```

```python
X = df3.drop(columns  = 'Response')
y = df3['Response']
```

```python
y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
ros = SMOTE()

Xr,yr = ros.fit_resample(X,y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.preprocessing import MinMaxScaler
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(Xr,yr)

scaler = MinMaxScaler()

scaler.fit(xtrain)
xtrainScaled = scaler.transform(xtrain)
xtestScaled = scaler.transform(xtest)

model_svm = SVC(kernel = 'poly' , class_weight={0:1 , 1:1.1})

model_svm.fit(xtrainScaled,ytrain)
```

```python
print(classification_report(ytrain , model_svm.predict(xtrainScaled)))
```

```python
print(classification_report(ytest , model_svm.predict(xtestScaled)))
```
