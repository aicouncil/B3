**Handling null values**
  - Remove
    - remove rows - it shouldn't be more then 1% of total data samples
    - remove columns - if majority of values are missing in a data columns or it is insignificalnt data columns (domain understanding or statistical analysis)
  - Impute - if majority of data values are known, so we can impute using mean or median

**Imputing**
   - Impute using mean  - if we dont have outliers
   - Impute using meadian - if we have outliers
   - Forward fill or backward fill

```python
#In age column 19.9% values are missing but approximately 80% values are known.
#Using the known values we can caluculate mean and fill the missing values
#To fill the missing values - fillna()

print(df['Age'].mean())
print(df['Age'].median())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
#'Outlet Location Type' is not biased but 'Outlet_Type' is biased
# So we can use information generated with respect to 'Outlet_location_type' to fill the missing values in 'Outlet_Size'
```

```python
#Function that will check Outlet location type and return 'Small' or 'Medium'
def fill_outlet_size(outlet_loc_type):
  if outlet_loc_type in ['Tier 1', 	'Tier 2']:
    return 'Small'
  else:
    return 'Medium'
```

```python
df.head(3)
```

```python
#fill only the missing values
df['Outlet_Size'] = df['Outlet_Size'].fillna(df['Outlet_Location_Type'].apply(fill_outlet_size))
```

**Standard Deviation & Histogram**

```python
iw = df['Item_Weight']
print(len(iw))
```

```python
total_samples = len(iw)
total_known_samples = total_samples - df['Item_Weight'].isna().sum()
print(total_known_samples)
```

```python
df['Item_Weight'].isna().sum()
```

```python
iw = df['Item_Weight']

iw_mean = iw.mean()
iw_std = iw.std()
```

```python
#%ge of data within 1st standard deviation
print(((iw > (iw_mean - iw_std)) & (iw < (iw_mean + iw_std))).sum()/total_known_samples)
```

```python
#%ge of data within 2nd standard deviation
print(((iw > (iw_mean - 2*iw_std)) & (iw < (iw_mean + 2*iw_std))).sum()/total_known_samples)
```

```python
#%ge of data within 3rd standard deviation
print(((iw > (iw_mean - 3*iw_std)) & (iw < (iw_mean + 3*iw_std))).sum()/total_known_samples)
```

```python
plt.hist(iw)

plt.axvline(iw_mean , c='black')

plt.axvline(iw_mean + iw_std , c = 'green')     #lower fence of 1st standard deviation
plt.axvline(iw_mean - iw_std , c = 'green')       #upper fence of first standard deviation

plt.axvline(iw_mean + 2 * iw_std , c = 'yellow')     #lower fence of 2nd standard deviation
plt.axvline(iw_mean - 2 * iw_std , c = 'yellow')       #upper fence of 2nd standard deviation

plt.axvline(iw_mean + 3 * iw_std , c = 'red')     #lower fence of 3rd standard deviation
plt.axvline(iw_mean - 3 * iw_std , c = 'red')       #upper fence of 3rd standard deviation

plt.show()
```

```python
#Any value more then uf_iv or lesser then lf_iv will be replaced with 'uf_iv' and 'lf_iv'
df['Item_Visibility'] = df['Item_Visibility'].clip(upper = uf_iv , lower = lf_iv)
```
