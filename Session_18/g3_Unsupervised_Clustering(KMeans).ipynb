**Elbow method** - To decide optimum number of clusters

```python
d3
```

```python
losses = []

for i in range(1,13):
  modelK = KMeans(n_clusters = i)
  modelK.fit(d3)
  #print(i, modelK.inertia_)
  losses.append(modelK.inertia_)
```

```python
plt.plot(range(1,13) , losses , marker = '*')
plt.show()
```

**Country data for clustering**
- HELP International have been able to raise around $ 10 million. Now the CEO of the NGO needs to decide how to use this money strategically and effectively. So, CEO has to make decision to choose the countries that are in the direst need of aid. Hence, your Job as a Data scientist is to categorise the countries using some socio-economic and health factors that determine the overall development of the country. Then you need to suggest the countries which the CEO needs to focus on the most.

```python
dfc = pd.read_csv('https://github.com/bipulshahi/Dataset/raw/refs/heads/main/Country-data.xls')
dfc.head()
```

```python
dfc1 = dfc.drop(columns = 'country')
dfc1.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(dfc1)
dfc2 = scaler.transform(dfc1)
```

```python
losses = []
for k in range(1,14):
  model_ct = KMeans(n_clusters=k)
  model_ct.fit(dfc2)
  losses.append(model_ct.inertia_)
```

**Elbow method**

```python
plt.plot(range(1,14) , losses , marker = '*')
plt.show()
```

**Silhouette scores**

```python
from sklearn.metrics import silhouette_score

X = dfc2.copy()
```

```python
silhouette_scores = []
for k in range(2,12):
  model_st = KMeans(n_clusters=k , random_state=0)
  model_st.fit(X)
  labels = model_st.predict(X)
  score = silhouette_score(X , labels)
  silhouette_scores.append(score)
```

```python
plt.plot(range(2,12) , silhouette_scores  , marker = '*')
plt.show()
```

```python
model_country = KMeans(n_clusters=4)
model_country.fit(dfc2)
```

```python
cluster_centers = pd.DataFrame(model_country.cluster_centers_, columns = dfc1.columns)
cluster_centers
```

```python
cluster_centers.plot.bar(figsize = (12,4))
```
