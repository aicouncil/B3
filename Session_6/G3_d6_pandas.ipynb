```python
df.head()
```

**Bivariate analysis for two categorical columns**

```python
#How many passengers suvived from each passenger class
pd.crosstab(df['Pclass'] , df['Survived'])
```

```python
pd.crosstab(df['Pclass'] , df['Survived']).plot.bar()
```

```python
#How many male and female passengers survived
pd.crosstab(df['Sex'] , df['Survived'])
```

```python
pd.crosstab(df['Sex'] , df['Survived']).plot.bar()
```

```python
#How many male and feamle passengers travelling in each class
pd.crosstab(df['Sex'], df['Pclass'])
```

```python
pd.crosstab(df['Sex'], df['Pclass']).plot.bar()
```

**Bivariate analysis for two continous columns**

```python
df.head()
```

```python
df[['Age', 'Fare']].corr()
```

```python
x1 = df['Age']
x2 = df['Fare']

n1 = ((x1 - x1.mean()) * (x2 - x2.mean())).sum()

d1 = (((x1 - x1.mean())**2).sum() * ((x2 - x2.mean())**2).sum())**0.5

c1 = n1/d1
print(c1)
```

**Multivariate Analysis**
   - Analyse considering multiple columns at a time

```python
df.head()
```

```python
#Average fare paid by 'male' and 'female' from each Passenger class
df.groupby(['Pclass' , 'Sex'])['Fare'].mean()
```

```python
df.groupby(['Pclass' , 'Sex'])['Fare'].mean().unstack()
```

```python
df.groupby(['Pclass' , 'Sex'])['Fare'].mean().unstack().plot.bar()
```

```python
#average age of survivals and non-survivals from each passenger class
df.groupby(['Pclass' , 'Survived'])['Age'].mean()
```

```python
df.groupby(['Pclass' , 'Survived'])['Age'].mean().unstack()
```

```python
df.groupby(['Pclass' , 'Survived'])['Age'].mean().unstack().plot.bar()
```
