* **Adjusted r-squared** - Adjusted R-squared is a modified version of R-squared that adjusts for the number of predictors in a regression model. It provides a more accurate measure of the model's goodness of fit by penalizing the inclusion of irrelevant predictors, unlike R-squared which always increases when more predictors are added. Therefore, adjusted R-squared is preferred when evaluating model fit, especially when dealing with multiple independent variables

```python
r2_score_train = r2_score(ytrain , ytrainP)
r2_score_test =  r2_score(ytest , ytestP)
```

```python
print(r2_score_train)
print(r2_score_test)
```

```python
xtrain.shape
```

```python
n_train = xtrain.shape[0]
n_test = xtest.shape[0]

p_train = xtrain.shape[1]
p_test = xtest.shape[1]

adjusted_r2_train = 1 - (1 - r2_score_train) * (n_train - 1) / (n_train - p_train - 1)
adjusted_r2_test = 1 - (1 - r2_score_test) * (n_test - 1) / (n_test - p_test - 1)

print(f"Adjusted R-squared (train): {adjusted_r2_train}")
print(f"Adjusted R-squared (test): {adjusted_r2_test}")
```

**feature Engineering**
  - Check for multicollinearity using VIF
  - The Variance Inflation Factor (VIF) is a measure used in statistics to detect multicollinearity among predictor variables in a regression model. It quantifies the severity of multicollinearity, which occurs when two or more predictor variables are highly correlated, potentially leading to unstable and unreliable regression coefficient estimates.

```python
import seaborn as sns
plt.figure(figsize = (25,25))
sns.heatmap(df2.corr() , annot =True , cmap = 'RdYlGn' , fmt = '.2f')
plt.show()
```

```python
X_vif = X[['Item_Weight', 'Item_Fat_Content', 	'Item_Visibility',  	'Item_MRP',  	'Outlet_Identifier' ,
           'Outlet_Establishment_Year', 'Outlet_Size', 	'Outlet_Location_Type']]
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
```

```python
#statsmodels requires a constant (intercept) for VIF calculation
Xs = sm.add_constant(X_vif)
Xs.head(2)
```

```python
vif_item_weight = variance_inflation_factor(Xs.values , 0)
vif_Item_Fat_Content = variance_inflation_factor(Xs.values , 1)
vif_item_visibility = variance_inflation_factor(Xs.values , 2)
vif_Item_mrp = variance_inflation_factor(Xs.values , 3)

print(vif_item_weight, vif_Item_Fat_Content, vif_item_visibility, vif_Item_mrp)
```

```python
vif_data = pd.DataFrame()
vif_data['feature'] = Xs.columns
vif_data['VIF'] = [variance_inflation_factor(Xs.values , i) for i in range(len(Xs.columns))]

vif_data
```
